```python
import numpy as np 
import matplotlib.pyplot as plt
# sklearn for linear regression and metrics 
# mean squared error and r2 score for evaluation 
from sklearn.linear_model  import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Load the dataset
# x represtents the number of hours studied and y represents the score obtained
X = np.array([[1], [2], [3], [4], [5]])
# y represents the scores obtained
y = np.array([50, 55, 65, 70, 75])
#ccomment 
```

```python
# Create a linear regression model
model = LinearRegression()
# Fit the model
model.fit(X, y)
# Make predictions
predictions = model.predict(X)
```

```python
#  mse  see if the error is big or small error 
mse = mean_squared_error(y, predictions)
print('mean_squared_error:', mse)

# Calculate R^2 score , see if the model is getting the relation correctly or not 
r2 = r2_score(y, predictions)
print('r2_score:', r2)
```

```python
# Plot the results
plt.scatter(X, y, color='blue', label='Actual Data')

# Plot the regression line
plt.plot(X, model.predict(X),linewidth= 2 , color='red', label='Regression Line')

# Add labels and title
plt.title('study hours vs score')
plt.xlabel('Hours Studied')
plt.ylabel('Score Obtained')

# Add legend
plt.legend()

#show the plot
plt.show()
```
